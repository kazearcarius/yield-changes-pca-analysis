```python

import numpy as np

# Set a random seed for reproducibility
np.random.seed(0)

# Define the parameters:
n_samples = 100   # Number of data points (observations)
n_variables = 5   # 5 yield change variables
mean = 0          # Mean close to 0
std_dev = 0.01    # A small standard deviation (yield changes)

# Generate the data: each column is an independent, uncorrelated yield change series
data = np.random.normal(loc=mean, scale=std_dev, size=(n_samples, n_variables))
```

```python
from sklearn.decomposition import PCA

# Create a PCA object (using the covariance matrix by default)
pca = PCA()

# Fit the PCA model on the data
pca.fit(data)

# Retrieve the proportion of variance explained by each component
explained_variance = pca.explained_variance_ratio_
print("Explained variance ratio for each component:")
for i, var in enumerate(explained_variance, 1):
    print(f"Component {i}: {var:.4f}")
```

When performing PCA on the dataset of five uncorrelated yield changes, we observe that each principal component explains a nearly equal portion of the total variance. This is expected since the variables were generated to have the same mean, standard deviation, and to be uncorrelated. In an ideal scenario, each of the five components would capture approximately 20% of the overall variance. Slight differences in the explained variance ratios occur due to sampling randomness, but no single component dominates the variance explanation.

```python
import matplotlib.pyplot as plt

# Create the scree plot
plt.figure(figsize=(6, 4))
plt.plot(range(1, len(explained_variance) + 1), explained_variance,
         marker='o', linestyle='--', color='b')
plt.title("Scree Plot: Variance Explained by Each Principal Component")
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.xticks(range(1, len(explained_variance) + 1))
plt.grid(True)
plt.show()
```
